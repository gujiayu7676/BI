```python
import pandas as pd
import mlxtend.frequent_patterns
from mlxtend.frequent_patterns import association_rules
import efficient_apriori 
```

```python
df = pd.read_csv('D:\文档\个人\AI课程\BI\code\L2-code\L2code\MarketBasket\Market_Basket_Optimisation.csv',header = None)
df
```

# 方法1：用efficient_apriori包中的apriori进行关联分析

```python
transactions = []
for i in range(len(df)):
    row = set(df.loc[i].dropna())
    transactions.append(row)
print(transactions)
```

```python
itemsets,rules = efficient_apriori.apriori(transactions,min_support = 0.05,min_confidence =0.3)
print("频繁项集：", itemsets)
print("关联规则：", rules)
```

# 方法2:使用mlxtend包

```python
data = []
for i in range(df.shape[0]):
    row = ["/".join(list(df.loc[i].dropna()))]
    data.append(row)
data     
```

```python
df1 = pd.DataFrame(data=data,columns=['basket'])
df1
```

```python
basket_hot_encoded= df1.basket.str.get_dummies('/')
basket_hot_encoded
```

```python
itemsets = mlxtend.frequent_patterns.apriori(basket_hot_encoded,use_colnames=True,min_support=0.05)
print("频繁项集：", itemsets)
```

```python
rules = association_rules(itemsets,metric='lift',min_threshold = 1)
rules =  rules.sort_values(by ='lift',ascending=False)
print("关联规则：", rules)
```
