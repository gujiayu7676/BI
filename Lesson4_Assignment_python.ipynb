### 2. 机器学习和函数/模块的思想有哪些异同？

你的回答：相同：都需要给定输入，有些函数无输入参数，也可以算作一种特别的输入叭。
不同：函数需要明确的解决问题的逻辑，我们可以用代码写出解决问题的方法。不要给输出。
机器学习不需要我们知道解决问题的逻辑，可以根据我们给的输入以及输出，进行自我学习，自己探索解决问题的方法。
